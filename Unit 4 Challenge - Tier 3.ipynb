**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Consider calling this function create_price_ratio.

You'd want this function to:
1. Take a filter of dfg, specifically where this filter constrains the London_Borough, as an argument. For example, one admissible argument should be: dfg[dfg['London_Borough']=='Camden'].
2. Get the Average Price for that Borough, for the years 1998 and 2018.
4. Calculate the ratio of the Average Price for 1998 divided by the Average Price for 2018.
5. Return that ratio.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018.

Bear in mind: you don't have to write a function like this if you don't want to. If you can solve the brief otherwise, then great! 

***Hint***: This section should test the skills you acquired in:
- Python Data Science Toolbox - Part One, all modules

```python
# Function create_price_ratio that gives a ratio by avg. price in 2018 / avg. price in 1998
def create_price_ratio(d):
    
    price_1998 = float(d['Average_Price'][d['Year'] == 1998])
    price_2018 = float(d['Average_Price'][d['Year'] == 2018])
    
    ratio_2018_1998 = [price_2018 / price_1998]
    
    return ratio_2018_1998
    
```

```python
# Create an empty dictionary to store borough:ratio_2018_1998
final_dict = {}
```

```python
# Use a forLoop to iterate through create_price_ratio function
for b in new_properties_final_avg['London Borough'].unique():
    each_borough = new_properties_final_avg[new_properties_final_avg['London Borough'] == b]
    final_dict[b] = create_price_ratio(each_borough)
print(final_dict)
```

```python
# Create a new variable to save the result as DataFrame
new_ratios= pd.DataFrame(final_dict)
```

```python
# Show top 5 records

new_ratios.head()
```

```python
# Transpose columns and rows

new_ratios_t = new_ratios.transpose()
new_ratios_t.head()
```

```python
# Reset indices

new_ratios_t = new_ratios_t.reset_index()
new_ratios_t.head()
```

```python
# Change Column Headings

new_ratios_t_revised = new_ratios_t.rename(columns={'index':'London_Borough',0:'Ratio'})
new_ratios_t_revised.head()
```

```python
# Sort Ratio in descending order
new_ratios_t_revised_desc = new_ratios_t_revised.sort_values(by='Ratio',ascending=False)
new_ratios_t_revised_desc.head()
```

```python
# Plot a bar graph
ax = new_ratios_t_revised_desc[['London_Borough','Ratio']].plot(kind='bar')
```
